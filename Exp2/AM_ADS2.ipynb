```python
#import libraries
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("/content/Tamilnadu agriculture yield data - Tamilnadu agriculture yield data.csv")
data.head()  #initial dataset
```

```python
data.shape
```

```python
print(data.isna().sum())
```

```python
from sklearn.impute import SimpleImputer
cat=SimpleImputer(missing_values=np.nan,strategy='most_frequent')  #for categorical variables
num=SimpleImputer(missing_values=np.nan,strategy='median')    #for numerical variables
```

```python
#handle state name
data["State_Name"]=cat.fit_transform(data[["State_Name"]])
print(data.isna().sum())
```

```python
#handle state name
data["District_Name"]=cat.fit_transform(data[["District_Name"]])
print(data.isna().sum())
```

```python
#handle crop year
data["Crop_Year"]=cat.fit_transform(data[["Crop_Year"]])
print(data.isna().sum())
```

```python
#handle season
data["Season"]=cat.fit_transform(data[["Season"]])
print(data.isna().sum())
```

```python
#handle crop
data["Crop"]=cat.fit_transform(data[["Crop"]])
print(data.isna().sum())
```

```python
#handle production
data["Production"]=num.fit_transform(data[["Production"]])
print(data.isna().sum())
```
